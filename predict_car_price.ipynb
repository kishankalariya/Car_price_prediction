## multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = mdata[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

```python
gdata= mdata.drop(["Year"],axis=1)
gdata=gdata.drop(['Model'],axis=1)
gdata=gdata.drop(['Price'],axis=1)

gdata.head()
```

## create dummy variable 

```python
dmdata=pd.get_dummies(gdata,drop_first=True)
```

```python
dmdata.head()
```

## Rearrange a bit

```python
dmdata.columns.values
```

```python
cols = ['log price', 'Mileage', 'EngineV',  'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_processed = dmdata[cols] 
data_processed.head()
```

##  linear_regression model

```python
y=data_processed['log price']
x1= data_processed.drop(['log price'],axis=1)
```

## scale the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x1)
```

```python
x = scaler.transform(x1)
```

## train test split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

## create regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

```python
reg.score(x_train,y_train)
```

```python
a=reg.coef_
```

```python
reg.intercept_
```

```python
fm = pd.DataFrame()
fm['feature']= x1.columns.values
fm['weight']= a
fm
```

## Testing

```python
y_hat_test= reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)

plt.xlabel("targets")
plt.ylabel("prediction")

plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test),columns=['prediction'])
df_pf
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test=y_test.reset_index(drop=True)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['dif']=df_pf['Target']-df_pf['prediction']
df_pf
```

```python
df_pf['Difference%'] = np.absolute(df_pf['dif']/df_pf['Target']*100)
df_pf
```
